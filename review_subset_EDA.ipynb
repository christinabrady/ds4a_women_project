```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Exploratory Data Analysis

## Overview

```python
review = pd.read_json("yelp_team7_dataset_review.json")
review
```

```python
review.columns
```

```python
review.dtypes
```

```python
review.isnull().sum()
```

```python
review['text length'] = review['text'].apply(len)
review.head()
```

```python
review.describe()
```

```python
review_cor = review[['stars', 'cool', 'useful', 'funny','text length']].corr()
fig = plt.figure(figsize=(12,8))
axes = sns.heatmap(review_cor, cmap='coolwarm', linewidth=1, linecolor='white', annot=True)
axes.set_title('Heatmap of Variables', fontsize=30)
```

- There's a strong positive correlation between **cool** and **useful** 
- There's a moderate positive correlation between **cool** and **funny**
- There's a moderate positive correlation between **funny** and **useful**


```python
fig = plt.figure(figsize=(10,8))
axes1 = sns.countplot(x='stars', data=review)
axes1.set_title('Stars')
axes1.set_ylabel('Count')
```

# By Year

```python
group_year=review.groupby(review['date'].dt.year)

reviews=group_year['review_id'].count()
useful = group_year['useful'].sum()
funny = group_year['funny'].sum()
cool = group_year['cool'].sum()

yearly_analysis=pd.DataFrame({"Number of reviews":reviews,
                        "Useful": useful,
                         "Funny":funny,
                         "Cool":cool
                        })
yearly_analysis

yearly_analysis.loc[2019:2022]
```

# By Business

```python
business_group= review.groupby("business_id")
rev_per_business = business_group["review_id"].count()
usefulness = business_group["useful"].sum()
coolness = business_group["cool"].sum()
funniness= business_group["funny"].sum()
business_analysis=pd.DataFrame({"Number of reviews":rev_per_business,
                                
                        "Usefulness":usefulness,
                         "Coolness":coolness,
                         "Funniess":funniness
                        })

sort_num_rev= business_analysis.sort_values("Number of reviews",ascending=False)
sort_num_rev
```

# Group by year and business

```python
year_business=review.groupby([review['date'].dt.year,"business_id"]).agg({"review_id": 'count',
                                                            "useful":'sum',
                                                           "cool":'sum',
                                                           "funny":'sum'})

```

```python
year_business
```

```python
year_business.loc[2019].sort_values('review_id',ascending=False).head(5)
```

# Pre-covid & during covid

# During Covid

```python
covid = review[(review['date']>='2020-01-01')&(review['date']<'2021-01-01')]
covid.head(2)
```

```python
fig = plt.figure(figsize=(8,8))
axes1 = sns.countplot(x='stars', data=covid)
axes1.set_title('Stars')
axes1.set_ylabel('Count')
```

```python
year_business=covid.groupby([covid['date'].dt.year,"business_id"]).agg({"review_id": 'count',
                                                                         "useful":'sum',
                                                                          "cool":'sum',
                                                                          "funny":'sum'})
year_business.sort_values("review_id",ascending=False).head(5)
```
